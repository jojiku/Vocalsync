```python
from faster_whisper import WhisperModel
import sys
import torch
import torchaudio as ta
from chatterbox.mtl_tts import ChatterboxMultilingualTTS
from IPython.display import Audio, display, clear_output
```

# TTS

```python
multilingual_model = ChatterboxMultilingualTTS.from_pretrained(device='cuda')
print("Model loaded successfully.")
```

```python
text1 = "В общем, чем больше я наблюдаю за вселенной, тем меньше верю, что в ней твариится хоть что-то доброе."
text2 = "Вот смотри, если твайё представление о создателе — что это кто-то всемогу́щий и совершенно добрый. А это довольно частое сочетание качеств, которыми описывают создателя."
text3 = "А я вот смотрю на катастрофы, которые обрушиваются на Землю и на всио живое. Вул каны, ура ганы, смерчи, землетрясения, болезни, эпидемии, различные уродства." 
text4 = "Посмотришь на весь этот список того, как жизнь превращается в ад на Земле по естественным причинам."
```

```python
AUDIO_PROMPT_PATH = "/kaggle/input/wetgwet/neil-audio_iMKlXbKX.wav"

wav = multilingual_model.generate(text4, 
                                  audio_prompt_path=AUDIO_PROMPT_PATH, 
                                  language_id="ru", 
                                  exaggeration=0.8, 
                                  cfg_weight=0.9,
                                  temperature=0.8,
                                  repetition_penalty=1.5,
                                  top_p=0.95,
                                  min_p=0.1
)

ta.save("test-ru.wav", wav, multilingual_model.sr)

display(Audio(wav.squeeze().numpy(), rate=multilingual_model.sr, autoplay=True))
print("Voice is Done!")
```
